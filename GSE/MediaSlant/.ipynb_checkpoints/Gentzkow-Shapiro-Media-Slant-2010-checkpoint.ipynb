# Media Slant and Text Analysis

This notebook aims to replicate parts of the text analysis performed in the seminal 2010 paper written by Matthew Gentzkow and Jesse Shapiro and published in Econometrica. A link to the article can be found [here](https://web.stanford.edu/~gentzkow/research/biasmeas.pdf).

The paper analyzes the selective use of **idologically-charged language** in media and its connection with political preferences by readers and owners alike. Crucially for us, in it the authors use natural language processing techniques hitherto scarcely used by economists. In particular, Gentzkow and Shapiro construct measures of slant based on language similarity between the lexicon used by over 400 national newspapers in the USA and the one used by politicians both sides of the aisle.

The **methodology** used by the authors can be best summarized as follows: 

- Consider official speeches by US congresspeople
- Identify the bi-grams and tri-grams most representative of Republicans and Democrats
- Compute the relative frequency of Democratic and Republican expressions in the articles published in over 400 newspapers
- Define the slant of each newspaper with respect to politicians
- Test whether slant is driven by consumers' vs. owners' preferences

The main **findings** can be succinctly summed up as:
- Slant highly correlates with political leaning of potential readers
- Identity of media owner does not explain much

```python
# Pre-processing steps
remove_punct = str.maketrans('','','!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~“”’')
combined_sw = stopwords.words('english')
combined_sw.append('speaker')
stop_words   = set(combined_sw)
stemmer      = SnowballStemmer('english')
```

```python
# Pre-processing function
def normalize_text(doc):
    """Takes documents as input and outputs clean tokens"""
    
    doc        = doc.replace('</br>', ' ')
    lower      = doc.lower()                                             # all lower case
    nopunc     = lower.translate(remove_punct)                           # remove punctuation
    words      = nopunc.split()                                          # split into tokens
    nostop     = [w for w in words if w not in stop_words]               # remove stopwords
    no_numbers = [w if not w.isdigit() else '#' for w in nostop]         # normalize numbers
    stem_text  = [stemmer.stem(w) for w in no_numbers]                   # stem each word
    
    return stem_text
```

```python
# Create list to fill
speeches = []

for doc in speech_df['speech']:
    
    raw_wip = normalize_text(doc)
    speeches.append(raw_wip)
```

```python
speech_clean = [" ".join(text) for text in speeches] #Convert each speech string-list to an actual string
print(type(speech_clean), len(speech_clean), type(speech_clean[0]), speech_clean[0])

# Create additional column in the speeches DF that is the pre-processed speeches
speech_df["preproc_speech"] = speech_clean
```

```python
# Define boolean mask for whether the speaker is a member of the Republican party
mask = (speech_df['party_'] == 'R')
#print(mask)

#Compute the chi-2 statistic between the bi-grams and tri-grams and the party affiliation of congressmembers
chi2score = chi2(data_tfidf, mask)[0]                         # Get the first array returned by the chi2 function, i.e. the chi-2 statistics of each association

print(type(chi2score), len(chi2score), type(chi2score[0]))    # 1,002,384 bi-grams and tri-grams in total
print(chi2score[:48])
```

```python
# Get the top chi-2
wscores = list(zip(tvec.get_feature_names(), chi2score)) # Create a list of (phrase, chiscore) tuples
print(len(wscores))

wchi2   = sorted(wscores, key=lambda x:x[1]) #Sort the list according to the chiscore
topchi2 = list(zip(*wchi2[-500:])) #Retrieve the top 500 words. Returns a list of 2 tuples. The first tuple contains the top 500 words, the second tuple lists the associated chi2 score.

#Create a DF with the top 500 phrases and their respective chi2 score.
top_phrases = pd.DataFrame({'phrase':list(topchi2[0]), 'chi2':list(topchi2[1])})
top_phrases.to_csv(datadir + 'top_500_speech_phrases.csv', index=False)
```

## Pre-processing the newspaper articles

```python
outlets = pd.read_stata(datadir + 'newspaper_outlets.dta')
print(outlets.shape)
outlets.head()
```

```python
# Preprocess articles in the NewsLibrary dataset that are from newspaper outlets only (do this year by year in order not to overload RAM)
year = 2009
useful_columns = ['n', 'date', 'year', 'source', 'title', 'wordcount', 'preproc_nl_articles', 'NPNAME1', 'APintI']

for filename in tqdm(os.listdir(np_data)):
    str_year = str(year)
    print(filename)
    df_ = pd.read_excel(np_data + filename)
    df_.drop('A', axis = 1, inplace = True)
    
    #Merge to the outlets DF
    df_ = pd.merge(df_, outlets, on = 'source', how = 'inner')
    print(df_.shape)
    
    #About 0.1% of the DF has NANS for text or title so replace these by empty strings
    df_['text'] = df_['text'].fillna('')
    df_['title'] = df_['title'].fillna('')

    #Create a column for the full text of the article: title + text
    df_['alltext'] = df_['title'] + ' ' + df_['text']

    # Define pre-processing list and perform it
    speeches = []

    for doc in df_['alltext']:

        raw_wip = normalize_text(doc)
        speeches.append(raw_wip)

    full_text = [" ".join(text) for text in speeches]
    print("Pre-processing done for year ", str_year)

    df_["preproc_nl_articles"] = full_text
    df_['year'] = str_year
    df_['n'] = 1
    print('Shape of DF of year ', str_year, ' is ', df_.shape)
    df_.loc[:,useful_columns].to_csv(datadir + 'NL_processed_np/NL_processed_np_' + str(year) + '.csv', index = False)
    year += 1
```

```python
# Load the data processed above
df_processed_np = pd.read_csv(datadir + 'NL_processed_np/NL_processed_np_2009.csv')

for i in range(2010, 2018):
    print('Concat with year:', i)
    df_ = pd.read_csv(datadir + 'NL_processed_np/NL_processed_np_' + str(i) + '.csv')
    
    df_processed_np = df_processed_np.append(df_, ignore_index=True)

df_processed_np.to_csv(datadir + 'NL_processed_np/NL_processed_np_allyears.csv')
```

## Media Slant

We will now compute a immigration-specific measure of media slant of NewsLibrary articles released yearly. The objective is to replicate the method developed by Gentzkow and Shapiro (2010).

The process below can be summarized in a series of steps:

1 - Of the 500 phrases that are most predictive of the speakers' affiliation to the Republican party, the code keeps the ones encountered in NewsLibrary articles at least 10 times (these result in 369 phrases)

2 - For each phrase _p_ and each congressperson _c_, the code computes the
	relative frequency of the phrase in Congressional speech

3 - The code then regresses the relative frequency on an indicator for the 
	congressperson's party, obtaining phrase-specific intercept and slope 
	coefficients _a_ and _b_
    
		- # Regresions   = # Top Phrases
		- # Observations = # Congresspeople

4 - Next, the code computes the relative frequency of each phrase in NewsLibrary
	articles released in a given year.

5 - Finally, for each newspaper we regress its relative immigrant-related phrase frequency (substracting _a_) on all phrase-specific _b_ obtained above (measures
	the phrase's ideological load). The resulting coeficient estimates reflect
	the time-varying measure of slant.

### First Step: Filter phrases

```python
tops = []

for i in tqdm(top_phrases['phrase']):
    
    freq = df_processed_np.preproc_nl_articles.str.count(i).sum()
    if freq > 10:
        tops.append([i,freq])

tops = pd.DataFrame(tops,columns = ['phrase','freq'])
```

```python
top_phrases = top_phrases.merge(tops, on = 'phrase', how = 'inner')
```

```python
top_phrases
```

### Second Step: Frequencies in Congresspeople Speeches

```python
speeches = speech_df.copy()
keep = ['speaker','party','preproc_speech']
speeches = speeches[keep]
speeches['party'] = pd.get_dummies(speeches['party'].astype('category'))

speeches.head()
```

```python
speeches['preproc_speech'] = speeches.groupby(['speaker','party'])['preproc_speech'].transform(lambda x: ' '.join(x))
speeches = speeches[['speaker','party','preproc_speech']].drop_duplicates().reset_index(drop = True)
for phrase in tqdm(top_phrases['phrase']):
    speeches[phrase] = speeches['preproc_speech'].str.count(phrase)

speeches.loc[:,'total'] = speeches.sum(axis=1)
speeches.iloc[:,3:] = speeches.iloc[:,3:].div(speeches.total,axis=0)
speeches = speeches.dropna(axis = 0).drop('total', axis=1)
speeches.head()
```

### Third Step: Obtain phrase-specific $\alpha$ and $\beta$

```python
coefficients = []

x = speeches['party'].values.reshape(-1,1)

for phrase in tqdm(top_phrases['phrase']):

    regr = LinearRegression()
    
    regr.fit(x, speeches[phrase])
    
    alpha = regr.intercept_
    beta  = regr.coef_[0]
    
    coefficients.append([phrase,alpha,beta])
    
coefficients = pd.DataFrame(coefficients, columns=['phrase','alpha','beta'])
```

### Fourth Step: Frequencies in NewsPapers

```python
newspapers = df_processed_np.copy()
keep = ['year','NPNAME1','preproc_nl_articles']
newspapers = newspapers[keep]
```

```python
newspapers['preproc_nl_articles'] = newspapers.groupby(['NPNAME1','year'])['preproc_nl_articles'].transform(lambda x: ' '.join(x))
newspapers = newspapers[['year','NPNAME1','preproc_nl_articles']].drop_duplicates().reset_index(drop = True)

for phrase in tqdm(top_phrases['phrase']):
    newspapers[phrase] = newspapers['preproc_nl_articles'].str.count(phrase)
    
newspapers.loc[:,'total'] = newspapers.drop('year', axis=1).sum(axis=1)

newspapers.iloc[:,3:] = newspapers.iloc[:,3:].div(newspapers.total,axis=0)
newspapers = newspapers.dropna(axis = 0).drop('total', axis=1)

newspapers.head()
```

### Fifth Step: Obtaining a time-varying measure of media slant 

```python
y = newspapers[(newspapers['NPNAME1'] == 'abilenereporternewsTX')&(newspapers['year'] == 2009)]
```

```python
np_coeff = newspapers.drop('preproc_nl_articles', axis=1)
np_coeff = pd.melt(np_coeff, id_vars=['NPNAME1','year'], var_name='phrase')

np_coeff.head()
```

```python
y = np_coeff[(np_coeff['NPNAME1'] == 'abilenereporternewsTX')&(np_coeff['year'] == 2009)]
```

```python
year_newspaper_slant = []

for newspaper in tqdm(list(np_coeff.NPNAME1.unique())):
    
    for year in range(2009, 2018):
        
        temp_df = np_coeff[(np_coeff['NPNAME1'] == newspaper) & (np_coeff['year'] == year)]
        
        if len(temp_df) == 0:
            pass
        else:
            freq  = temp_df['value'].values.reshape(-1,1)
            alpha = coefficients['alpha'].values.reshape(-1,1)

            y     = freq - alpha
            beta  = coefficients['beta'].values.reshape(-1,1)

            regr.fit(beta, y)

            slant = regr.coef_[0][0]

            year_newspaper_slant.append(tuple([year,newspaper,slant]))

```

```python
np_slant = pd.DataFrame(year_newspaper_slant, columns = ['year','newspaper','slant'])
```
